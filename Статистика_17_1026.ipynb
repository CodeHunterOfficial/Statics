3.6 Применение регрессионного анализа и интерпретация результатов

Регрессионный анализ используется для изучения отношений между зависимыми и независимыми переменными. Он может быть полезен для прогнозирования значений зависимой переменной на основе значений независимых переменных.

Результаты регрессионного анализа могут быть интерпретированы с помощью коэффициентов регрессии. Коэффициенты регрессии представляют собой числа, которые указывают на изменение зависимой переменной при изменении независимой переменной на единицу.

Для нахождения коэффициентов регрессии b0 и b1 в уравнении $y = b_0 + b_1x$ можно использовать метод наименьших квадратов (МНК).

Шаги для нахождения коэффициентов регрессии:

Соберите данные: Соберите значения зависимой переменной (y) и независимой переменной (x) для каждого наблюдения в выборке.

Вычислите средние значения: Вычислите средние значения ȳ и x̄ для зависимой и независимой переменных соответственно.

Вычислите отклонения: Для каждого наблюдения вычислите отклонения от средних значений: $(y - ȳ)$ и $(x - x̄).$

Вычислите произведения отклонений: Умножьте каждое отклонение $(y - ȳ)$ на соответствующее отклонение $(x - x̄)$ для каждого наблюдения.

Вычислите коэффициенты регрессии: Используя формулы, коэффициенты регрессии могут быть вычислены следующим образом:
https://sun.tsu.ru/mminfo/2016/Dombrovski/book/chapter-2/chapter-2-2.htm

$$b_1 =\frac{ \sum_{n=1}^{n} (y_i-\overline y)(x_i-\overline x)}{\sum_{n=1}^{n} (x_i-\overline x)^2} $$

$$b_0 = \overline y - b_1 \overline x$$

Таким образом, вы можете найти коэффициенты регрессии b0 и b1 для уравнения $y = b_0 + b_1x$, используя метод наименьших квадратов.



Коэффициент детерминации (R-квадрат) также является важным показателем регрессионного анализа. Он указывает на то, какая доля вариации в зависимой переменной может быть объяснена независимыми переменными. Значение R-квадрат может быть от 0 до 1, где 1 означает, что все вариации в зависимой переменной могут быть объяснены независимыми переменными.

Формула:

$$R=1-\frac{ \sum_{i=1}^{2}(y_i-\check y_i)}{\sum_{i=1}^{2}(y_i-y)^2}=\frac{SSres}{SStot}$$

где

$y_i, \check y_i$- фактические и расчётные значения объясняемой переменной.

SStot (общая сумма квадратов отклонений) - это сумма квадратов разностей между каждым наблюдаемым значением и средним значением выборки.

SSres (Sum of Squares Residuals) - это сумма квадратов остатков, т.е. разниц между фактическими значениями зависимой переменной и теми, которые были предсказаны моделью.

Интерпретация значения R-квадрат:

* Ближе к 1: Высокое значение R-квадрат указывает на то, что модель хорошо соответствует данным и обеспечивает хорошее объяснение вариации в зависимой переменной.
* Ближе к 0: Низкое значение R-квадрат означает, что модель плохо соответствует данным и не объясняет много вариации в зависимой переменной.

Однако следует отметить, что R-квадрат не дает информации о причинно-следственных связях между переменными и не может быть использован для сравнения моделей с разным набором независимых переменных.

Важно помнить, что регрессионный анализ не может доказать причинно-следственные связи между переменными, а только указать на связь между ними.

Пример.

Допустим, у нас есть следующие данные:

| Независимая переменная | Зависимая переменная |
|------------------------|----------------------|
|           1            |          2           |
|           2            |          4           |
|           3            |          5           |
|           4            |          4           |
|           5            |          6           |

Сначала нужно вычислить среднее значение зависимой переменной:

mean(y) = (2 + 4 + 5 + 4 + 6) / 5 = 4.2

Затем нужно вычислить SStot:

SStot = Σ(yᵢ - ȳ)² = (2 - 4.2)² + (4 - 4.2)² + (5 - 4.2)² + (4 - 4.2)² + (6 - 4.2)² = 5.6

Теперь нужно построить линейную регрессионную модель, чтобы вычислить предсказанные значения зависимой переменной. Для этого нужно вычислить коэффициенты регрессии a и b:

b = Σ((xᵢ - x̄)(yᵢ - ȳ)) / Σ(xᵢ - x̄)² = (1-3.0)*(2-4.2) + (2-3.0)*(4-4.2) + (3-3.0)*(5-4.2) + (4-3.0)*(4-4.2) + (5-3.0)*(6-4.2) / ((1-3.0)² + (2-3.0)² + (3-3.0)² + (4-3.0)² + (5-3.0)²) = 0.9

a = ȳ - bx̄ = 4.2 - 0.9*3.0 = 1.5

Таким образом, уравнение линейной регрессии будет выглядеть как y = 1.5 + 0.9x

Теперь можно вычислить предсказанные значения зависимой переменной (y) для каждого значения независимой переменной (x) с помощю формулу $\check y=1.5 + 0.9x$:

| Независимая переменная | Зависимая переменная | Предсказанное значение |
|------------------------|----------------------|-----------------------|
|           1            |          2           |          2.4          |
|           2            |          4           |          3.3          |
|           3            |          5           |          4.2          |
|           4            |          4           |          5.1          |
|           5            |          6           |          6.0          |

Теперь нужно вычислить SSres:

SSres = Σ(yᵢ - ŷᵢ)² = (2 - 2.4)² + (4 - 3.3)² + (5 - 4.2)² + (4 - 5.1)² + (6 - 6.0)² = 1.7

И, наконец, можно вычислить R²:

R² = 1 - (SSres / SStot) = 1 - (1.7 / 5.6) ≈ 0.696

Задача: Построить линейную модель, которая наилучшим образом объясняет зависимость между уровнем образования и показателем бедности. Для этого необходимо рассчитать коэффициенты модели - коэффициент сдвига (пересечение с осью Y) и коэффициент наклона (угол наклона прямой).

Гипотеза: Определить, насколько хорошо наша модель действительно объясняет поведение показателя бедности. Для этого мы можем вычислить коэффициент детерминации (R²), который показывает, какой процент дисперсии объясняется нашей моделью.

Гипотеза: Проверить, отличается ли коэффициент наклона (B₁) от нуля. Если коэффициент B₁ не равен нулю, это означает наличие взаимосвязи между уровнем образования и показателем бедности.

Задача: Предсказать будущие значения показателя бедности на основе данных о независимой переменной (уровень образования). Мы будем использовать регрессионную модель для предсказания и проверки гипотез.

Шаг 1. Построить диаграмму рассеяния

```python
import pandas as pd
import matplotlib.pyplot as plt

# Загрузка данных из CSV файла
data = pd.read_csv('http://d396qusza40orc.cloudfront.net/statistics/lec_resources/states.csv')

# Построение скаттерплота
plt.scatter(data['poverty'], data['hs_grad'])
plt.xlabel('hs_grad')
plt.ylabel('poverty')
plt.title('Scatter Plot of Murder vs Assault')
plt.show()

# Вывод описательных статистик
descriptive_stats = data.describe()
print(descriptive_stats)
```

```python
import numpy as np
# Разделение данных на независимую переменную (X) и зависимую переменную (y)
X = data['hs_grad'].values
y = data['poverty'].values

# Вычисление средних значений X и y
mean_X = np.mean(X)
mean_y = np.mean(y)

# Вычисление разностей от средних значений X и y
diff_X = X - mean_X
diff_y = y - mean_y

# Вычисление коэффициента наклона b1
b1 = np.sum(diff_X * diff_y) / np.sum(diff_X**2)

# Вычисление коэффициента пересечения с осью y (b0)
b0 = mean_y - b1 * mean_X

# Вывод коэффициентов
print("Коэффициент b0:", b0)
print("Коэффициент b1:", b1)
```

image.png

```python
# Построение скаттерплота с линией регрессии
plt.scatter(X, y)
plt.plot(X, b0 + b1*X, color='red')
plt.xlabel('hs_grad')
plt.ylabel('poverty')
plt.title('Scatter Plot with Regression Line')
plt.show()
```

Коэффициент корреляции (R) показывает, насколько сильно связаны две переменные между собой. Его значение может находиться в диапазоне от -1 до 1, где -1 означает полную обратную связь, 0 - отсутствие связи, а 1 - полную прямую связь.

```python
# Расчет коэффициента корреляции (R)
R = data['poverty'].corr(data['hs_grad'])
# Вывод коэффициента корреляции
print("Коэффициент корреляции (R):", R)
```

Коэффициент детерминации (R²) показывает, какую долю дисперсии зависимой переменной можно объяснить независимой переменной при помощи линейной регрессии. Его значение также может находиться в диапазоне от 0 до 1, где 0 означает, что независимая переменная не объясняет вариацию зависимой переменной, а 1 - что объясняет ее полностью.

```python
# Вычисление прогнозных значений y
y_pred = b0 + b1 * X

# Вычисление суммы квадратов отклонений реальных значений y от их среднего значения
SS_total = np.sum((y - mean_y)**2)

# Вычисление суммы квадратов отклонений прогнозных значений y от их среднего значения
SS_res = np.sum((y_pred - mean_y)**2)

# Вычисление коэффициента детерминации R²
R_squared = 1 - (SS_res / SS_total)

# Вывод коэффициента детерминации
print("Коэффициент детерминации R²:", R_squared)
```

Значение R² = 0.4422 говорит о том, что около 44.22% дисперсии зависимой переменной может быть объяснено независимой переменной при помощи линейной регрессии. Это означает, что существует умеренная связь между этими двумя переменными, и около 44.22% вариации в зависимой переменной может быть предсказано или объяснено независимой переменной "hs_grad". Оставшиеся 55.78% вариации могут быть обусловлены другими факторами или иными переменными, которые не включены в модель.

*** На заметку: Коэффициент детерминации (R²) и коэффициент корреляции (R) - это два разных показателя, которые используются для измерения степени взаимосвязи между двумя переменными.***

```python
import matplotlib.pyplot as plt

# Получение прогнозных значений
y_pred = b0 + b1 * X

# Вычисление остатков
residuals = y - y_pred

# Построение графика остатков
plt.scatter(y_pred, residuals)
plt.axhline(y=0, color='r', linestyle='--')
plt.xlabel('Прогнозные значения')
plt.ylabel('Остатки')
plt.title('График остатков')
plt.show()
```

График остатков отражает изменчивость остатков на протяжении всего времени, что позволяет оценить гомоскедастичность (однородность дисперсии).

Гомоскедастичность в контексте статистики означает однородность дисперсии остатков модели на разных уровнях независимой переменной. То есть, при гомоскедастичности, разброс остатков остается постоянным по всем значениям независимой переменной.

Гомоскедастичность важна для правильной интерпретации и оценки модели. Если остатки модели гетероскедастичны (имеют неравномерную дисперсию), это может привести к некорректным статистическим выводам и неверным интерпретациям результатов.

Наличие гомоскедастичности позволяет использовать классические статистические методы, такие как метод наименьших квадратов (OLS), для оценки параметров модели и проведения статистических тестов. Когда гомоскедастичность нарушена, могут использоваться корректирующие методы, например, взвешенный метод наименьших квадратов (WLS) или робастные стандартные ошибки, чтобы получить более точные и надежные результаты.

Таким образом, гомоскедастичность важна для обеспечения надежности и корректности статистических выводов и интерпретации результатов модели.

```python
import statsmodels.api as sm
import matplotlib.pyplot as plt

# Вычисление остатков
residuals = y - y_pred

# Построение QQ-Plot
sm.qqplot(residuals, line='s')
plt.xlabel('Теоретические квантили')
plt.ylabel('Квантили остатков')
plt.title('QQ-Plot остатков')
plt.show()
```

# Предсказание значение

Предсказание в регрессионном анализе основывается на линии тренда, которая также называется регрессионной прямой. Линия тренда позволяет предсказывать значения зависимой переменной на определенном уровне независимой переменной. Например, мы можем использовать регрессионный анализ, чтобы выяснить, как уровень образования влияет на процент людей, проживающих в бедности в штате.

Для этого мы строим регрессионную прямую, которая проходит через наши выборочные данные. Если наша выборка соответствует генеральной совокупности, то мы можем использовать регрессионный анализ для предсказания значений зависимой переменной на основе линии тренда.

Чтобы сделать предсказание, мы можем нанести точку на регрессионную прямую и получить самую точную оценку значения зависимой переменной на определенном уровне независимой переменной. Но мы также можем использовать уравнение линии регрессии для расчета предсказанных значений. Например, если мы хотим узнать, какой процент людей будет жить в бедности при 95% уровне образования населения, мы можем подставить значение 95 в уравнение линии регрессии и получить предсказанное значение показателя бедности.

Однако важно помнить, что предсказание имеет свои ограничения. Например, линия тренда может принимать отрицательные значения, что не имеет смысла для некоторых показателей. Кроме того, линейная связь между переменными может быть далеко не всегда очевидной и может изменяться на разных промежутках значений независимой переменной.

Пример. Допустим, мы хотим изучить взаимосвязь между ростом и возрастом человека на промежутке от 0 до 20 лет. В начале мы можем наблюдать сильную линейную зависимость между этими переменными, но затем значение роста перестанет изменяться после достижения значения 100. Однако, если в нашей выборке присутствуют только люди младше 20 лет, то линия тренда продолжит расти в бесконечность.

Поэтому использование линейной регрессии для предсказания значений имеет свои ограничения. Прежде чем использовать регрессионный анализ для предсказания значений, необходимо убедиться в надежности и адекватности модели. Для этого нужно проверить выполнение нескольких условий: линейности взаимосвязи, нормальности распределения остатков и гомоскедастичности.

Если нарушены эти условия, то регрессионная модель может потерять свою предсказательную силу. Также важно помнить, что линейная зависимость между двумя переменными в выборке не означает, что эта зависимость будет сохраняться на всех уровнях значений переменных.

Таким образом, при использовании линейной регрессии для предсказания значений необходимо учитывать ограничения модели и проверять ее надежность и адекватность.
